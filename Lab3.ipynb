```python
# Порядковий номер студента
n = 28

# Розмір матриць
matrix_size = n + 2

# Створення матриць
matrix_A = np.random.randint(0, 10, size=(matrix_size, matrix_size))
matrix_B = np.random.randint(0, 10, size=(matrix_size, matrix_size))

print("Матриця A:")
print(matrix_A)

print("Матриця B:")
print(matrix_B)
```

```python
# Додавання
matrix_sum = matrix_A + matrix_B
print("Сума матриць:")
print(matrix_sum)

# Віднімання
matrix_diff = matrix_A - matrix_B
print("Різниця матриць:")
print(matrix_diff)

# Ділення
# Заміна ділення на 0 на 1
matrix_div = matrix_A / matrix_B
matrix_div[matrix_div == np.inf] = 1  # Заміна NaN на 1
print("Ділення матриць:")
print(matrix_div)
```

```python
# Скалярне множення
scalar = 5
matrix_scalar_mult = scalar * matrix_A
print("Скалярне множення на", scalar, ":")
print(matrix_scalar_mult)

# Векторне множення
matrix_vector_mult = matrix_A @ matrix_B
print("Векторне множення:")
print(matrix_vector_mult)
```

```python
# Визначник
matrix_det = np.linalg.det(matrix_A)
print("Визначник матриці A:", matrix_det)

# Власні числа
matrix_eigvals, matrix_eigvecs = np.linalg.eig(matrix_A)
print("Власні числа матриці A:")
print(matrix_eigvals)

print("Власні вектори матриці A:")
print(matrix_eigvecs)
```

```python
# Задана точка для обчислення похідної
x_point = np.pi / 5

# Визначення похідної
def f_derivative(x):
    return 2 / (1 + 4 * x**2) + 5 / (1 + x**2) - 1 / x

# Обчислення похідної в заданій точці
derivative = f_derivative(x_point)
print("Похідна в точці", x_point, ":", derivative)
```

```python
# Заданий проміжок інтегрування
a = -2 * np.pi
b = 2 * np.pi

# Визначення первісної
def f_integral(x):
    return 2 * np.arctan(2 * x) + 5 * np.arctan(x) + np.log(np.abs(x))

# Обчислення первісної на заданому проміжку
integral = f_integral(b) - f_integral(a)
print("Первісна на проміжку [", a, ";", b, "]:", integral)
```

```python
# Визначення функції для оптимізації
def f_optimize(x):
    return math.atan(2 * x) + 5 * math.atan(x) - np.log(np.abs(x))  # Заміна arcctg на math.atan

# Знаходження глобального мінімуму
global_min_result = optimize.minimize(f_optimize, x0=0, method="Nelder-Mead")
global_min_x = global_min_result.x[0]
global_min_y = f_optimize(global_min_x)

print("Глобальний мінімум:", global_min_x, global_min_y)

# Знаходження локального мінімуму
local_min_result = optimize.minimize(f_optimize, x0=0.5, method="BFGS")
local_min_x = local_min_result.x[0]
local_min_y = f_optimize(local_min_x)

print("Локальний мінімум:", local_min_x, local_min_y)
```
